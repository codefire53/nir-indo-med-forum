# Unknown Percentage

## A. Tydi

```python
gold_data = []
actual_data = []

with open('../Results/syifa_results_full_tydi.json', 'r') as actual_out:
    actual_data = json.load(actual_out)

with open('../Dataset/Questions/question-syifa-test.json', 'r') as gold_out:
    gold_data = json.load(gold_out)
```

```python
gold_ids = []
neg_ids = []
for row in gold_data:
    row_gold_ids = []
    row_neg_ids = []
    for row_ctx in row['ctxs']:
        row_gold_ids.append(row_ctx['id'])
    for row_ctx in row['neg_ctxs']:
        row_neg_ids.append(row_ctx['id'])
    gold_ids.append(row_gold_ids)
    neg_ids.append(row_neg_ids)
```

```python
total_cnt = 0
unk_cnt = 0
unks = []
avg_unk = 0
for idx, row in enumerate(actual_data):
    unk_per_question = 0
    total_cnt += len(row['ctxs'])
    for ctx in row['ctxs']:
        if ctx['id'] not in gold_ids[idx] and ctx['id'] not in neg_ids[idx]:
            unk_cnt += 1
            unk_per_question += 1
    unks.append(unk_per_question/len(row['ctxs'])*100)
    avg_unk += unk_per_question/len(row['ctxs'])*100
avg_unk/=len(actual_data)

print(f'Unknown percentage: {unk_cnt/total_cnt*100}')
print(f'Average unknown per question percentage: {avg_unk}')
print(f'Median unknown per question percentage: {sorted(unks)[int(len(unks)//2)]}')
print(f'Min unknown per question percentage: {min(unks)}')
print(f'Max unknown per question percentage: {max(unks)}')
sns.distplot(unks)
```

# Unknown Pairs Sampling for Annotation

```python
gold_data = []


with open('../Dataset/Questions/question-syifa-test.json', 'r') as gold_out:
    gold_data = json.load(gold_out)
```

```python
import csv
corpus_map = dict()

with open('../Dataset/Corpus/corpus-syifa-normalized.tsv', 'r') as f:
    reader = csv.reader(f, delimiter='\t')
    for row in reader:
        pid = row[0]
        text = row[1]
        title = row[2]
        corpus_map[pid] = {
            'text': text,
            'title': title
        }
```

```python
gold_ids = []
neg_ids = []
question_map = dict()
for row in gold_data:
    row_gold_ids = []
    row_neg_ids = []
    qid = row['question']['id']
    title = row['question']['title']
    text = row['question']['text']
    question_map[qid] = {
        'title': title,
        'text': text
    }
    for row_ctx in row['ctxs']:
        row_gold_ids.append(row_ctx['id'])
    for row_ctx in row['neg_ctxs']:
        row_neg_ids.append(row_ctx['id'])
    gold_ids.append(row_gold_ids)
    neg_ids.append(row_neg_ids)
```

```python
len(gold_ids)
```

```python
import glob 
import json

qp_unk_map = dict()
for filename in tqdm(glob.glob('../Results/*.json')):
    if 'dpr' not in filename.lower():
        continue
    json_data = []
    with open(filename, 'r') as f:
        json_data = json.load(f)
    for idx,row in enumerate(json_data):
        qid = row['question']['id']
        for ctx in row['ctxs']:
            if ctx['id'] not in gold_ids[idx] and ctx['id'] not in neg_ids[idx]:
                composite_id = f"{qid}_{ctx['id']}"
                if composite_id not in qp_unk_map:
                    qp_unk_map[composite_id]= 0
                qp_unk_map[composite_id] += 1
qp_unk_map = sorted(qp_unk_map.items(), key=lambda x: x[1], reverse=True)
```

```python
import pandas as pd

unknown_annotation_df = pd.DataFrame()
for composite_id, freq in tqdm(qp_unk_map):
    splitted_id =  composite_id.split('_')
    qid = splitted_id[0]
    pid = splitted_id[1]
    unknown_annotation_df = unknown_annotation_df.append({
        'question_id': qid,
        'question_text': question_map[qid]['text'],
        'question_title': question_map[qid]['title'],
        'passage_id': pid,
        'passage_text': corpus_map[pid]['text'],
        'passage_title': corpus_map[pid]['title'],
        'frequency': freq
    }, ignore_index=True)
```

```python
import numpy as np

sorted_df = unknown_annotation_df.groupby('question_id').agg({'frequency':np.median}).reset_index().sort_values(['frequency'],ascending=False)
```

```python
selected_index = sorted_df['question_id'][:15]
```

```python

sorted_qid = [int(qid) for qid in selected_index]
sorted_qid = sorted(sorted_qid)
sorted_qid = [str(qid) for qid in sorted_qid]
```

```python
final_df = pd.DataFrame()
for qid in sorted_qid:
    df_per_qid = unknown_annotation_df[unknown_annotation_df['question_id']==qid][:10]
    final_df = final_df.append(df_per_qid, ignore_index=True)
final_df.head()
```
